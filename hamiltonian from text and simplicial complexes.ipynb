### ChatGPT
Here is an example code that can be used to extract a knowledge graph from text and plot the knowledge graph with node labels for entities using spaCy and NetworkX libraries:

The following code installs the necessary packages when run in terminal:

````python
import spacy
import networkx as nx
import matplotlib.pyplot as plt

# Load the spaCy model
nlp = spacy.load('en_core_web_sm')

# Define the text to extract the knowledge graph from
text = "Barack Obama was the 44th President of the United States. He served two terms from 2009 to 2017. He is married to Michelle Obama."

# Use spaCy to process the text and extract entities
doc = nlp(text)

# Create an empty graph
G = nx.Graph()

# Add nodes to the graph, with the node labels being the entities in the text
for ent in doc.ents:
    G.add_node(ent.text)

# Add edges to the graph, connecting entities that appear close to each other in the text
for ent1 in doc.ents:
    for ent2 in doc.ents:
        if ent1.text == ent2.text:
            continue
        if ent1.end <= ent2.start:
            G.add_edge(ent1.text, ent2.text)

# Plot the knowledge graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()
````

This code uses spaCy to process the text and extract entities, and uses NetworkX to create a graph with nodes being the entities and edges being connections between entities that appear close to each other in the text. The graph is then plotted using matplotlib, with node labels being the entity names and edge labels being the connections between entities.

Here is an example of code to extract and plot a labeled knowledge graph from text using spaCy and NetworkX:

````python
import spacy
import networkx as nx
import matplotlib.pyplot as plt

# Load the spaCy model
nlp = spacy.load('en_core_web_sm')

# Define the text to extract the knowledge graph from
text = "Barack Obama was the 44th President of the United States. He served two terms from 2009 to 2017. He is married to Michelle Obama."

# Use spaCy to process the text and extract entities
doc = nlp(text)

# Create an empty graph
G = nx.Graph()

# Add nodes to the graph, with the node labels being the entities in the text
for ent in doc.ents:
    G.add_node(ent.text)

# Add edges to the graph, connecting entities that appear close to each other in the text
for ent1 in doc.ents:
    for ent2 in doc.ents:
        if ent1.text == ent2.text:
            continue
        if ent1.end <= ent2.start:
            G.add_edge(ent1.text, ent2.text, label=ent1.label_ + "-" + ent2.label_)

# Plot the knowledge graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()
````

This code uses spaCy to extract entities from the text, then creates a NetworkX graph where the nodes are the entities and the edges are connections between entities that appear close to each other in the text. The edge labels are the relationship between the entities, which is represented by the entity labels in spaCy. Finally, the knowledge graph is plotted using matplotlib.
